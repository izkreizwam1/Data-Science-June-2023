```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.ndimage import convolve
```

### create image with clear vertical edge 

```python
edgy = np.array(8*[[0,0,0,10,10,0,0,0]])
plt.imshow(edgy)
```

### create Vertical filter

```python
filter_V = np.array([[1,1,1],[0,0,0],[-1,-1,-1]])
filter_V
```

### try to detect edges 

```python
edge_v = convolve(edgy, filter_V)
plt.imshow(edge_v)
```

```python
edge_v
```

### no edges were detected!

### create horizontal filter

```python
filter_H = np.array([[1,0,-1],[1,0,-1],[1,0,-1]])
```

```python
filter_H
```

```python
edge_h = convolve(edgy, filter_H)
plt.imshow(edge_h)
```

### Edges were detected , but... WHY are they called horizontal 

```python
edge_h
```

Using dot product

```python
e_v = np.ones([6,6])
for i in range(5):
    for j in range(5):
        matr = edgy[i:i+3,j:j+3].dot(filter_V)
        e_v[i,j] = sum(sum(matr))
e_v
```

```python
plt.imshow(e_v)
```
